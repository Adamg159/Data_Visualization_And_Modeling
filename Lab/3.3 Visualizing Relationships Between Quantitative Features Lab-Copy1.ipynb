```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

df = pd.read_csv('Employee_productivity.csv')
df.head()
```

```python
sns.regplot(x='YearsAtCompany', y='ProductivityScore', data=df)
plt.title('Productivity Score by Years At Company')
plt.show()
```

```python
X = sm.add_constant(df['YearsAtCompany'])
y = df['ProductivityScore']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
sns.regplot(x='HoursWorked', y='ProductivityScore', data=df)
plt.title('Productivity Score by Hours Worked')
plt.show()
```

```python
X = sm.add_constant(df['HoursWorked'])
y = df['ProductivityScore']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
sns.regplot(x='FocusScore', y='ProductivityScore', data=df)
plt.title('Productivity Score by FocusScore')
plt.show()
```

```python
X = sm.add_constant(df['FocusScore'])
y = df['ProductivityScore']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
sns.regplot(x='TeamMeetingsPerWeek', y='ProductivityScore', data=df)
plt.title('Productivity Score by Team Meetings Per Week')
plt.show()
```

```python
X = sm.add_constant(df['TeamMeetingsPerWeek'])
y = df['ProductivityScore']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
df['Department'].value_counts()
```

```python
df['Department'] = df['Department'].map({'Marketing': 0, 'Sales': 1, 'Engineering': 2, 'HR': 3})

X = sm.add_constant(df['Department'])
y = df['ProductivityScore']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
sns.regplot(x='Department', y='ProductivityScore', data=df)
plt.title('Productivity Score by Department')
plt.show()
```

```python
df['DeskLocation'].value_counts()
```

```python
df['DeskLocation'] = df['DeskLocation'].map({'Remote': 0, 'Open Plan': 1, 'Private Office': 2})

X = sm.add_constant(df['DeskLocation'])
y = df['ProductivityScore']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
sns.regplot(x='DeskLocation', y='ProductivityScore', data=df)
plt.title('Productivity Score by Desk Location')
plt.show()
```

There seems to be no close to or no linear relationship with every variable in the data besides the Focus Score. So the only thing that employees should be working on if they want to be more productive is working on their focus.

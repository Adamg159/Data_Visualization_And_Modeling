The data that I'm using is the race data from formula 1 for the years 1950-2026. This data has been gathered and posted to the public domain by Chris Newell. It contains basically every stat regarding formula1 but I'm choosing just to cover race results and I added the drivers so that I can see who the drivers are and what races they were at that top the stats I'll be looking at. This dataset is interesting to me as I enjoy watching F1 and knowing more about something I enjoy can't ever be a bad thing.

1. Which drivers have driven the most laps in their career?
2. Which drivers have had the best average finishing position throughout their careers with atleast 50 race finishes? 
3. Which drivers have had the most pole positions (grid as 1)?
4. which constructors have had the most unique drivers race for them?
5. How many races have been won with a gap of less than 5 seconds?

```python
import pandas as pd
results_df     = pd.read_csv("results.csv")

#replace all \n null markers with NaN to make sure calculations work later
results_df.replace('\\N', pd.NA, inplace=True)
results_df['laps']         = pd.to_numeric(results_df['laps'],         errors='coerce')
results_df['position']     = pd.to_numeric(results_df['position'],     errors='coerce')
results_df['grid']         = pd.to_numeric(results_df['grid'],         errors='coerce')
results_df['milliseconds'] = pd.to_numeric(results_df['milliseconds'], errors='coerce')
results_df.head()
```

```python
constructors_df = pd.read_csv('constructors.csv')
constructors_df.head()
```

```python
drivers_df = pd.read_csv('drivers.csv')
drivers_df.head()
```

```python
races_df = pd.read_csv('races.csv')
races_df.head()
```

```python
mostlaps_df = results_df.groupby('driverId')['laps'].sum().reset_index()

mostlaps_df = mostlaps_df.merge(drivers_df[['driverId', 'forename', 'surname']], on='driverId')
mostlaps_df['name'] = mostlaps_df['forename'] + ' ' + mostlaps_df['surname']

mostlaps_df = mostlaps_df[['name', 'laps']].sort_values('laps', ascending=False)
top10_ids = mostlaps_df.head(10)['driverId']
mostlaps_df.head(10)
```

1.Which drivers have driven the most laps in their career?
The drivers who have driven the most laps in their career are Fernando Alonso with 23189 laps and Lewis Hamilton with 21765 laps.
I was curious how long this was in terms of time so I'm gonna also answer that question below this one.

```python
results_df['milliseconds'] = pd.to_numeric(results_df['milliseconds'], errors='coerce')
top10_time = results_df[results_df['driverId'].isin(top10_ids)].groupby('driverId')['milliseconds'].sum().reset_index()
top10_time = top10_time.merge(drivers_df[['driverId', 'forename', 'surname']], on='driverId')
top10_time['name'] = top10_time['forename'] + ' ' + top10_time['surname']
top10_time['total_hours'] = (top10_time['milliseconds'] / 3600000).round(2)

print(results_df[results_df['driverId'].isin(top10_ids)]['milliseconds'].isnull().groupby(results_df['driverId']).sum())
top10_time[['name', 'total_hours']].sort_values('total_hours', ascending=False)
```

1.5. How long have the drivers with the most laps spent on track?
It would seem the data is not super complete for this type of assessment but it's still somewhat interesting to look at regardless and Lewis having no missing data at 541 hours spent during races seems like a really long time to be under the load that a F1 car puts on a persons body.

```python
avgfin_df = results_df.dropna(subset=['position'])
avgfin_df = avgfin_df.groupby('driverId').agg(avg_pos=('position', 'mean'), races=('position', 'count')).reset_index()
avgfin_df = avgfin_df[avgfin_df['races'] >= 50]
avgfin_df = avgfin_df.merge(drivers_df[['driverId', 'forename', 'surname']], on='driverId')
avgfin_df['name'] = avgfin_df['forename'] + ' ' + avgfin_df['surname']
avgfin_df['avg_pos'] = avgfin_df['avg_pos'].round(2)
avgfin_df = avgfin_df[['name', 'avg_pos', 'races']].sort_values('avg_pos')
avgfin_df.head(10)
```

2.Which drivers have had the best average finishing position throughout their careers with atleast 50 race finishes? 
The leaders in this list are almost all new names from the most laps list with the exception of lewis hamilton which helps explain why most people consider him the greatest F1 driver of all time. This also supports the conclusion that you don't have to be dominant to have a long career in F1.

```python
mostpoles_df = results_df[results_df['grid'] == 1].groupby('driverId').size().reset_index(name='poles')
mostpoles_df = mostpoles_df.merge(drivers_df[['driverId', 'forename', 'surname']], on='driverId')
mostpoles_df['name'] = mostpoles_df['forename'] + ' ' + mostpoles_df['surname']
mostpoles_df = mostpoles_df[['name', 'poles']].sort_values('poles', ascending=False)
mostpoles_df.head(10)
```

3.Which drivers have had the most pole positions (grid as 1)?
Once again we see Lewis hamilton in the top 10. This list has alot of similarities with the average finishing position list which makes alot of sense as starting further up the grid would naturally lead to a better finishing position.

```python
mostdrivers_df = results_df.groupby('constructorId')['driverId'].nunique().reset_index(name='unique_drivers')
mostdrivers_df = mostdrivers_df.merge(constructors_df[['constructorId', 'name']], on='constructorId')
mostdrivers_df = mostdrivers_df[['name', 'unique_drivers']].sort_values('unique_drivers', ascending=False)
mostdrivers_df.head(10)
```

4.which constructors have had the most unique drivers race for them?
This one doesn't really show any sort of meaningful stat related to the performance of the teams but I think it's somewhat interesting to look at. It's cool to see Ferrari, Williams, and Mclaren on the list as well as these are the only 3 teams on this list that are still currently on the grid.

```python
winners_df    = results_df[results_df['positionOrder'] == 1][['raceId', 'milliseconds']].rename(columns={'milliseconds': 'winner_ms'})
runnerups_df  = results_df[results_df['positionOrder'] == 2][['raceId', 'milliseconds']].rename(columns={'milliseconds': 'p2_ms'})
closewins_df  = winners_df.merge(runnerups_df, on='raceId').dropna(subset=['winner_ms', 'p2_ms'])
closewins_df['gap_s'] = (closewins_df['p2_ms'] - closewins_df['winner_ms']) / 1000
closewins_df  = closewins_df[closewins_df['gap_s'] < 5]
closewins_df  = closewins_df.merge(races_df[['raceId', 'year', 'name']], on='raceId')
closewins_df['gap_s'] = closewins_df['gap_s'].round(3)
closewins_df  = closewins_df[['year', 'name', 'gap_s']].sort_values('gap_s')
print(closewins_df.count())
closewins_df.head(10)
```

5.How many races have been won with a gap of less than 5 seconds?
It would appear that the answer to this question is 374 which is alot more than I was expecting but I guess 5 seconds is a pretty big amount of time. The closest races are really interesting as well with 3 of them being decided by just one hundredth of a second which is crazy close.

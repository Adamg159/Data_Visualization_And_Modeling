# Lab Instructions

You are part of a data team for a financial institution that uses an AI model to determine who will and will not be granted a loan.  Recently, there have been complaints from customers that the model unfairly discriminates based on the applicant's age and sex. A court has ordered that the financial institution must provide information about how the model works.  

Data from recent loan applications is given in `loan_data.csv`.  The data dictionary is below:

| Feature              | Description | Data Type |
| :---------------- | :------: | ----: |
|person_age|Age of the person|Float
|person_gender|Gender of the person|Categorical
|person_education|Highest education level|Categorical
|person_income|Annual income|Float
|person_emp_exp|Years of employment experience|Integer
|person_home_ownership|Home ownership status (e.g., rent, own, mortgage)|Categorical
|loan_amnt|Loan amount requested|Float
|loan_intent|Purpose of the loan|Categorical
|loan_int_rate|Loan interest rate|Float
|loan_percent_income|Loan amount as a percentage of annual income|Float
|cb_person_cred_hist_length|Length of credit history in years|Float
|credit_score|Credit score of the person|Integer
|previous_loan_defaults_on_file|Indicator of previous loan defaults|Categorical
|loan_status (target variable)|Loan approval status: 1 = approved; 0 = rejected|Integer

Visualize the relationship between age, sex and 5 other features with `loan_status`.  Which features seem to drive the model's decision to grant a loan?  Which seem unrelated?  Should the financial institution be concerned about potential legal trouble?  Make sure to justify your answers with specific references to your visualizations.

**You must use quantitative visualizations (boxplots or histograms) when comparing quantitative features.  DO NOT categorize quantitative features.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('loan_data.csv')
df.head()
```

```python
sns.histplot(x='credit_score',data=df, hue='loan_status', element='step', palette='bright', stat='probability', common_norm=False, multiple='fill')
plt.title('Loan status by Credit Score (Normalized)')
plt.ylabel('Percentage of People')
plt.xlabel('Credit Score')
plt.show()
```

```python
sns.histplot(x='person_age',data=df, hue='loan_status', element='step', palette='bright')
plt.title('Loan status by Age')
plt.ylabel('Amount of People')
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(x='person_income',data=df, hue='loan_status', element='step', palette='bright', stat='probability', common_norm=False, multiple='fill')
plt.title('Loan status by Annual Income (Normalized)')
plt.ylabel('Percentage of People')
plt.xlabel('Annual Income')
plt.show()
```

```python
sns.histplot(x='person_gender',data=df, hue='loan_status', element='step', palette='bright', stat='probability', common_norm=False, multiple='fill')
plt.title('Loan status by Sex (Normalized)')
plt.ylabel('Percentage of People')
plt.xlabel('Sex')
plt.show()
```

```python
sns.histplot(x='loan_amnt',data=df, hue='loan_status', element='step', palette='bright', stat='probability', common_norm=False, multiple='fill')
plt.title('Loan status by Loan Amount (Normalized)')
plt.ylabel('Percentage of People')
plt.xlabel('Loan Amount')
plt.show()
```

```python
sns.histplot(x='loan_percent_income',data=df, hue='loan_status', element='step', palette='bright', stat='probability', common_norm=False, multiple='fill')
plt.title('Loan status by Percentage of Income (Normalized)')
plt.ylabel('Percentage of People')
plt.xlabel('Percentage of Income')
plt.show()
```

```python

sns.histplot(x='person_emp_exp',data=df, hue='loan_status', element='step', palette='bright')
plt.title('Loan status by Years of Employment Experience')
plt.ylabel('Amount of People')
plt.xlabel('Years of Employment Experience')
plt.show()
```

From my analysis, I would say that Percentage of Age(over retirement age), Income, Annual Income, Credit Score, and Loan Ammount are directly correlated to whether or not your loan will be accepted. On the other hand Sex, Years of Employment Experience seem to not make a difference. There is a chance that the institution could have legal trouble because of the 0% acceptance rate over a certain age. Although I don't see any other reason why somebody would have a case.

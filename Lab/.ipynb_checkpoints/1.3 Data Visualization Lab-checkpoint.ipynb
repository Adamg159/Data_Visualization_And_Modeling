```python
import pandas as pd

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
df.head()
```

```python
engineered_songs = df.loc[ (df['key'] == 'A') &
                           (df['bpm'] > 120) &
                           (df['bpm'] < 140)]
```

```python
vampire = df.loc[df['track_name'] == 'vampire']

vampire
```

```python
df['bpm'].plot(kind='hist', title='Most Common BPMs Among Top Spotify Songs in 2023', ylabel='Songs', xlabel='BPM')
```

```python
df['key'].value_counts().plot(kind='bar', title='Most Common Keys Among Top Spotify Songs in 2023', ylabel='Songs', xlabel='Key')
```

```python
df['mode'].value_counts().plot(kind='bar', title='Most Common Modes Among Top Spotify Songs in 2023', ylabel='Songs', xlabel='Mode')
```

```python
df['danceability_%'].plot(kind='hist', title="Most Common Danceability %'s Among Top Spotify Songs in 2023", ylabel='Songs', xlabel='Danceability %')
```

```python
df['energy_%'].plot(kind='hist', title="Most Common Energy %'s Among Top Spotify Songs in 2023", ylabel='Songs', xlabel='Energy %')
```

```python
df['speechiness_%'].plot(kind='hist', title="Most Common Speechiness %'s Among Top Spotify Songs in 2023", ylabel='Songs', xlabel='Speechiness %')
```

```python
df['acousticness_%'].plot(kind='hist', title="Most Common Acousticness %'s Among Top Spotify Songs in 2023", ylabel='Songs', xlabel='Acoutsticness %')
```

```python
engineered_songs = df.loc[ (df['key'] == 'C#') &
                           (df['bpm'] > 120) &
                           (df['bpm'] < 140) &
                           (df['mode'] == 'Major') &
                           (df['energy_%'] > 60) &
                           (df['energy_%'] < 80) &
                           (df['danceability_%'] > 65) &
                           (df['danceability_%'] < 80) &
                           (df['speechiness_%'] < 10) &
                           (df['acousticness_%'] < 20)]
engineered_songs.head()
```
